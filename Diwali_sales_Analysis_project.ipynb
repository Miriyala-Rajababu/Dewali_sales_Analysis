# Data analysis

### Gender

```python

# Define conditions
male = df["Gender"] == "M"
female = df["Gender"] == "F"

# Calculate counts of male and female
male_count = male.sum()
female_count = female.sum()


# Plotting counts of male and female
plt.bar(['Male', 'Female'], [male_count, female_count],color=["blue","orange"])

# Adding labels and title
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Count of Male and Female')
plt.text(0, male_count + 1, str(male_count), ha='center')  # x=0 for "Male" bar
plt.text(1, female_count + 1, str(female_count), ha='center')
plt.show()
```

#### From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men

## Total amount by Gendar 

```python
# Calculating the total amount for males and Female
male_total_amount = df[(df["Gender"] == "M")]["Amount"].sum()
Female_total_amount = df[(df["Gender"] == "F")]["Amount"].sum()
plt.bar(["Male", "Female"], [male_total_amount, Female_total_amount], color=["blue", "orange"])
plt.text(0,male_total_amount,str(male_total_amount),ha="center")
plt.text(1,Female_total_amount,str(Female_total_amount),ha="center")
plt.title("Total amount By Gender")
plt.show()
```

## Age

```python

# Age counts for different ranges
age_below_18 = df[df["Age"] < 18]["Age"].count()
age_20 = df[(df["Age"] > 18) & (df["Age"] <= 30)]["Age"].count()
age_30 = df[(df["Age"] > 30) & (df["Age"] <= 40)]["Age"].count()
age_40 = df[(df["Age"] > 40) & (df["Age"] <= 50)]["Age"].count()
age_50 = df[(df["Age"] > 50) & (df["Age"] <= 60)]["Age"].count()
age_60 = df[(df["Age"] > 60) & (df["Age"] <= 70)]["Age"].count()
age_70 = df[df["Age"] > 70]["Age"].count()

# Create a list of the age ranges and their corresponding counts
age_groups = ['Below_18', '18-30', '30-40', '40-50', '50-60', '60-70', 'Above-70']
age_counts = [age_below_18, age_20, age_30, age_40, age_50, age_60, age_70]

# Plotting the bar chart
bars = plt.bar(age_groups, age_counts)

# Adding labels and title
plt.xlabel('Age Groups')
plt.ylabel('Number of People')
plt.title('Age Distribution')

# Add count values on top of the bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 0.05, str(height), ha='center', va='bottom', fontweight='bold')

# Show the plot
plt.show()
```

## states

```python

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(sales_state['State'], sales_state['Orders'], color='skyblue')
plt.xlabel('State')
plt.ylabel('Total Orders')
plt.title('Total Orders by State')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()
```

```python
total_amount_state=df.groupby(["State"], as_index=False)["Amount"].sum().sort_values(by='Amount',ascending=False)
plt.bar(total_amount_state["State"],total_amount_state['Amount'])
plt.xticks(rotation=90) 
plt.title("total amount by state")
plt.show()
```

#### From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively

```python
Zone=df.groupby(['Zone'])['Orders'].count()
plt.bar(Zone.index, Zone.values)
plt.title("purchased by zone")
plt.xlabel("zone")
plt.ylabel("orders")
```
